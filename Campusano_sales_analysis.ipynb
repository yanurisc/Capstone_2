# EmporiUm Sales Analysis — Massachusetts & New York

This notebook analyzes sales data for two cross-region territories:
- Territory 1: Massachusetts, managed by Bo Heap (Northeast Region)
- Territory 2: New York, managed by See Ellefson (East Region)

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### StoreSales
This file contains all individual sales transactions including date, store, product, and sale amount.

```python
df_sales = pd.read_csv('StoreSales.csv')
df_sales.info()
```

### StoreDetail
This file contains information about each store including location, state, territory manager, and region.


```python
df_stores = pd.read_csv('StoreDetail.csv')
df_stores.info()
```

### Products
This file contains the product list including product numbers, names, and category IDs.

```python
df_products = pd.read_csv('Products.csv')
df_products.info()
```

### ProductCategories
This file contains the category and subcategory names linked to each product.

```python
df_categories = pd.read_csv('ProductCategories.csv')
df_categories.info()
```

### Customer_List
This file contains registered rewards customers including their ID, name, email, and phone number.

```python
df_customers = pd.read_csv('customer_list.csv', sep='|')
df_customers.info()
```

## Core Project Requirements

### Q1: Territory Overview
Who are the territory managers for the sales territories assigned? 
What are the store IDs and cities for the stores in each assigned sales territory?

```python
ma_stores = df_stores[df_stores['State'] == 'Massachusetts']
ny_stores = df_stores[df_stores['State'] == 'New York']

print("Territory 1 — Massachusetts")
print(ma_stores[['Store Location', 'Store ID', 'Territory Manager']].to_string(index=False))

print("\nTerritory 2 — New York")
print(ny_stores[['Store Location', 'Store ID', 'Territory Manager']].to_string(index=False))
```

```python
# Territory 1 Massachusetts has 18 stores across cities like Boston, Worcester, Lowell, and Nantucket (Store IDs 730 and 801–817), all managed by Bo Heap.
# Territory 2 New York has 12 stores across cities like Albany, Buffalo, Brooklyn, and New York City (Store IDs 840–851), all managed by See Ellefson.
```

### Q2: Monthly Revenue Trend
What is monthly total revenue for in-store sales in each of the two sales territories, over the full period covered by the data?

```python
# Convert Transaction Date to datetime
df_sales['Transaction Date'] = pd.to_datetime(df_sales['Transaction Date'])

# Create a YearMonth column
df_sales['YearMonth'] = df_sales['Transaction Date'].dt.to_period('M')

# Filter sales for each territory
ma_sales = df_sales[df_sales['Store ID'].isin(ma_stores['Store ID'])]
ny_sales = df_sales[df_sales['Store ID'].isin(ny_stores['Store ID'])]

# Group by month and sum revenue
ma_monthly = ma_sales.groupby('YearMonth')['Sale Amount'].sum().reset_index()
ny_monthly = ny_sales.groupby('YearMonth')['Sale Amount'].sum().reset_index()

print("Massachusetts Monthly Revenue")
print(ma_monthly.to_string(index=False))
print("\nNew York Monthly Revenue")
print(ny_monthly.to_string(index=False))
```

```python
# Massachusetts revenue goes from around $69,000/month in early 2022 all the way up to $277,000 in October 2025 — showing very strong consistent growth over the 4 years.
# New York revenue goes from around $58,000/month in early 2022 up to $185,000 in October 2025 — also growing strongly but at a lower level than Massachusetts.
# Both territories had a big jump starting in 2023 — revenue nearly doubled compared to 2022.
```

### Q3: Store Performance Ranking
How would you rank the sales performance of each store in each sales territory? Which are the top performing stores?

```python
# Group sales by store and sum revenue
store_revenue = df_sales.groupby('Store ID')['Sale Amount'].sum().reset_index()

# Merge with store details to get city names
store_revenue = store_revenue.merge(df_stores[['Store ID', 'Store Location', 'State']], on='Store ID')

# Filter for each territory and rank
ma_ranked = store_revenue[store_revenue['State'] == 'Massachusetts'].sort_values('Sale Amount', ascending=False)
ny_ranked = store_revenue[store_revenue['State'] == 'New York'].sort_values('Sale Amount', ascending=False)

print("Massachusetts Store Rankings")
print(ma_ranked[['Store Location', 'Store ID', 'Sale Amount']].to_string(index=False))
print("\nNew York Store Rankings")
print(ny_ranked[['Store Location', 'Store ID', 'Sale Amount']].to_string(index=False))
```

```python
# Massachusetts top 3 stores: 
# Worcester (StoreID 817) — $602,183
# Leominster (StoreID 807) — $338,009
# Nantucket (StoreID 810) — $335,548

# New York top 3 stores:
# New York City (StoreID 850) — $621,565
# Rochester (StoreID 847) — $607,148
# Poestenkill (StoreID 846) — $334,176
```

### Q4: Top Customers
Comparing the customer ID from the customer list data with the rewards ID from the sales data, who were the top customers in each sales territory?

```python
# Merge sales with customer list
sales_customers = df_sales.merge(df_customers, left_on='RewardsID', right_on='cust_id', how='left')

# Filter for each territory
ma_customers = sales_customers[sales_customers['Store ID'].isin(ma_stores['Store ID'])]
ny_customers = sales_customers[sales_customers['Store ID'].isin(ny_stores['Store ID'])]

# Group by customer and sum revenue
ma_top = ma_customers.groupby('name')['Sale Amount'].sum().reset_index().sort_values('Sale Amount', ascending=False).head(10)
ny_top = ny_customers.groupby('name')['Sale Amount'].sum().reset_index().sort_values('Sale Amount', ascending=False).head(10)

print("Top 10 Customers — Massachusetts")
print(ma_top.to_string(index=False))
print("\nTop 10 Customers — New York")
print(ny_top.to_string(index=False))
```

```python
# Massachusetts top 3 customers:
# Beverly Crusher — $4,758
# Tracy Jordan — $4,197
# Jesse Pinkman — $4,170

# New York top 3 customers:
# Julian Bashir — $6,039
# Jason Morgan — $3,918
# Mellie Grant — $3,805

# Julian Bashir appears in both territories, spending $3,883 in Massachusetts and $6,039 in New York, making him the highest spending rewards customer overall.
```

### Q5: Product Category Analysis
What is the number of transactions per month by product category in each assigned territory? What is total sales revenue per month by category? What might this tell you about the most popular products, and where could there be opportunity for growth?

```python
# Merge sales with products and categories
sales_products = df_sales.merge(df_products, on='Prod Num').merge(df_categories, on='CategoryID')

# Filter for each territory
ma_cat = sales_products[sales_products['Store ID'].isin(ma_stores['Store ID'])]
ny_cat = sales_products[sales_products['Store ID'].isin(ny_stores['Store ID'])]

# Group by category and sum revenue
ma_cat_revenue = ma_cat.groupby('Category')['Sale Amount'].agg(['sum', 'count']).reset_index()
ma_cat_revenue.columns = ['Category', 'Total Revenue', 'Transactions']
ma_cat_revenue = ma_cat_revenue.sort_values('Total Revenue', ascending=False)

ny_cat_revenue = ny_cat.groupby('Category')['Sale Amount'].agg(['sum', 'count']).reset_index()
ny_cat_revenue.columns = ['Category', 'Total Revenue', 'Transactions']
ny_cat_revenue = ny_cat_revenue.sort_values('Total Revenue', ascending=False)

print("Massachusetts — Revenue & Transactions by Category")
print(ma_cat_revenue.to_string(index=False))
print("\nNew York — Revenue & Transactions by Category")
print(ny_cat_revenue.to_string(index=False))
```

```python
# Both territories have the same category ranking:
# Technology & Accessories, highest revenue by far ($28.4M MA, $21.5M NY)
# Textbooks, second highest revenue but actually the most transactions in MA
# Art Supplies, third in revenue 
# Apparel and Merchandise, fourth in revenue
# Stationery and Supplies, most transactions in both territories
# Books (General) lowest revenue and transactions
```

```python
# Stationery and Supplies has the most transactions (99,209 in Massachusetts) but ranks 5th in revenue, meaning customers buy it often but in small amounts.
```
